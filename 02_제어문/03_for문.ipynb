```python
# for문의 기본구조

# for 변수 in 리스트(또는 튜플, 문자열):
#     실행할 문장1
#     실행할 문장2
#     ...

lis = [1, 2, 3]
```

```python
# 전형적인 for문

test_list = ["one", "two", "three"]

for i in test_list:
    print(i)
```

```python
# 다양한 for문의 사용

a = [(1, 2), (3, 4), (5, 6)]

for i, j in a:
    print(i+j)
```

```python
# for문의 응용

# 5명의 시험 점수
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1
    if score >= 60:
        print(f"id: {id}, 합격입니다.")
    else:
        print(f"id: {id}, 불합격입니다.")
```

```python
# for문과 continue문

scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1

    if score < 60:
        continue
    
    print(f"id가 {id} 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)
print(list(a))

b = range(1,10)
print(list(b))
```

```python
# 1부터 10까지 더하는 반복문

hap = 0

for i in range(1,11): # [1, 2, 3, ... , 10]
    hap += i

print(hap)
```

```python
# for와 range()를 이용하여 구구단을 만들어 보세요.
# 출력예시는
# 2 * 1 = 2
# 2 * 2 = 4
# ...
# 9 * 9 = 81

for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i} * {j} = {i * j}")
```

```python
# enumerate 함수 활용하기
fruits = ["apple", "banana","orange"]

for fruit in fruits:
    print(fruit)

for (i, fruit) in enumerate(fruits):
    print(f"{i} : {fruit}")
```

```python
# zip 함수로 여러 리스트 함께 순회하기
names = ["홍길동", "유재석", "강호동"]
scores = [85, 92, 80]

for name, score in zip(names, scores):
    print(f"{name} : {score}점")
```
