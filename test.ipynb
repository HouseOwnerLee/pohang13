````python
# 문제 2 
# 사용자로부터 숫자를 입력받아 해당 숫자의 구구단을 출력하는 프로그램을 작성하시오.

num = int(input("숫자를 입력: "))

for i in range(1,10):
    print(f"{num} * {i} = {num * i}")
````

````python
# 문제 4
# 두 개의 숫자를 입력받아 두 수 사이의 모든 정수의 합을 반환하는 함수 sum_between(a, b)를 작성하시오.
# (a와 b 포함)

def sum_between(a, b):
    hap = 0
    for num in range(a, b + 1):
        hap += num
    return hap

a = int(input("숫자 a 입력: "))
b = int(input("숫자 b 입력: "))

print(sum_between(a, b))
````

````python
# 문제 5
# 가변 인자를 받아 다음 정보를 계산하는 함수 calculate_stats(*numbers)를 작성하시오.

# 평균 (소수점 2자리)
# 최댓값
# 최솟값
# 개수

def calculate_stats(*numbers):
    hap = 0
    for num in numbers:
        hap += num
    
    print(f"평균 : {round(hap / len(numbers), 2)}")
    print(f"최댓값 : {max(numbers)}")
    print(f"최솟값 : {min(numbers)}")
    print(f"길이 : {len(numbers)}")

nums = list(map(int, input("숫자 입력: ").split(" ")))

calculate_stats(*nums)
````
