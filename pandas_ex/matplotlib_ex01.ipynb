# tick_params() 함수 : 눈금의 스타일을 다양하게 설정
- axis : 설정이 적용될 축을 지정
- direction : 'in', 'out'으로 설정하면 눈금이 안/밖으로 표시 {'in', 'out', 'inout'} 중 선택
- length : 눈금의 길이를 지정
- pad : 눈금과 레이블과의 거리를 지정
- labelsize : 레이블의 크기를 지정
- labelcolor : 레이블의 색상을 지정
- top/botton/left/right : True/False로 지정하면 눈금이 표시될 위치를 선택
- width : 눈금의 너비를 지정
- color : 눈금의 생삭을 지정

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.2)

plt.plot(x,x,'bo')
plt.plot(x,x**2,color='#e35f62',marker="*",linewidth=2)
plt.plot(x,x**3,color='springgreen', marker='^',markersize=9)
plt.xticks(np.arange(0,2.2,0.2),labels=['Jan','','Feb','','Mar','','May','','June','','July'])
plt.yticks(np.arange(0,7),('0','1GB', '2GB', '3GB', '4GB','5GB', '6GB'))
plt.tick_params(axis='x',direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')
plt.show()
```

# subplot() 함수

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False # 마이너스 부호가 깨지는 것을 방지함
x1 = np.linspace(start=0.0,stop=5.0,num=50)
x2 = np.linspace(0.0,2.0,num=50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle("Main Title") # 전체 제목

plt.subplot(2,1,1) # 2행 1열에서 첫번째
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2-1 subplots') # subplot에 대한 제목
plt.xlabel("time (m)")
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2) # 2행 1열에서 두번째
plt.plot(x2,y2,'r.-')
plt.title('A tale of 2-2 subplots') # subplot에 대한 제목
plt.xlabel("time (s)")
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

# contour 차트

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y **2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
XX, YY = np.meshgrid(x,y)
ZZ = f(XX,YY)
plt.title("Contour plots")
plt.contourf(XX,YY,ZZ, alpha=0.75, cmap='hot')
plt.contour(XX,YY,ZZ, colors = 'black')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = [70,90,80]
y2 = [60,70,90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5
xlabel = ['국어', '영어', '수학']
plt.title("Bar Chart")
plt.bar(x,y1,yerr=err1, ecolor='yellow', width=0.3,  color='red', align='center', label = '점수')
plt.bar(x+0.4,y2,yerr=err2, width=0.3, color='blue', align='center', label='반평균')
plt.xticks(x+0.4/2, xlabel)
plt.ylim(0,100)
plt.xlabel('과목')
plt.ylabel('점수')
plt.legend()
plt.show()
```

# pie 차트

```python
import matplotlib.pyplot as plt

labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45, 9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%', shadow = True,
       startangle = 90, counterclock=False)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops = {'width': 0.7, 'edgecolor':'w', 'linewidth': 5} # width : 부채꼴 너비, linewidth : 부채꼴 간격
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle = 260, counterclock=False, colors=colors,
       wedgeprops = wedgeprops)
plt.show()
```
