```python
def n_square(x):
    return x ** 2
```

```python
df4_1 = df["Age"].map(n_square) # 하나의 속성에 함수를 실행한 값을 반환하는 함수
df4_1
```

```python
def n_square(x, n):
    return x ** n
```

```python
df4 = df[["Age", "SibSp", "Fare"]].apply(n_square, n=2) # 여러개의 속성들에 함수를 실행한 값을 반환하는 함수
df4
```

### 객체 연산

```python
data = [[1,10,100],[2,20,200],[3,30,300]]
col = ['col1','col2','col3']
row = ['row1','row2','row3']
df = pd.DataFrame(data,row,col)
df
```

```python
result = df.add(1) # sub(), mul(), div(), mod()
result
```

```python
result = df + 1
result
```

```python
data2 = [[3],[4],[5]]
df2 = pd.DataFrame(data = data2, index = row,columns = ['col1'])
df2
```

```python
result = df.add(df2) # 같은 열끼리 계산함
result
```

```python
result = df.add(df2,fill_value = 0)
result
```
