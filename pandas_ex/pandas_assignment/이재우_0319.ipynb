```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
```

```python
# 1번
data = df['Survived'].value_counts()
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.bar(data.index, data, color=['red','green'], width = 0.7)
plt.xticks([0,1], labels=['사망', '생존'])
plt.xlabel('생존여부')
plt.ylabel('인원수')
plt.show()
```

```python
# 2번
data = df.groupby(['Survived', 'Pclass']).size().unstack()
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.bar(data.index, data[1], width = 0.7, label ='1등급')
plt.bar(data.index, data[2], bottom = data[1], width = 0.7, label ='2등급')
plt.bar(data.index, data[3], bottom = data[1] + data[2], width = 0.7,  label ='3등급')
plt.legend()
plt.xticks([0,1], labels=['사망', '생존'])
plt.xlabel('생존여부')
plt.ylabel('인원수')
plt.show()
```

```python
# 3번
data = df.groupby('Embarked').size()
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.title('탑승지역')
plt.pie(data, labels=data.index, autopct = '%1.2f%%')
plt.show()
```

```python
# 4번
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.hist(df['Age'], bins=[0,9,19,29,39,49,59,69,79,89], rwidth=1, color='blue', alpha=0.4, density=True)
plt.xticks([0,9,19,29,39,49,59,69,79,89])
plt.xlabel('나이')
plt.show()
```
