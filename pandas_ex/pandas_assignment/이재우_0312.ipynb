```python
# [문제1] 아래 코드를 활용해 주어진 문제에 대한 코드를 작성하시오.
import pandas as pd

data = [
    ["037730", "3R", 1510, 7.36],
    ["036360", "3SOFT", 1790, 1.65],
    ["005760", "ACTS", 1185, 1.28],
]
columns = ["종목코드", "종목명", "현재가", "등락률"]
df = pd.DataFrame(data=data, columns=columns)
df.set_index('종목코드', inplace=True)
df
```

```python
# (1) 데이터프레임에서 037730 종목의 등락률을 출력하세요.
df['등락률']['037730']
```

```python
# (2) 데이터프레임에서 iloc 속성을 사용하여 첫번째, 두번째 행을 슬라이싱하세요.
df.iloc[0:2]
```

```python
# (3) 데이터프레임에서 loc 속성을 사용하여 첫번째, 두번째 행을 슬라이싱하세요.
df.loc["037730":"036360"]
```

```python
# (4) 데이터프레임에서 iloc 속성을 사용하여 첫 번째 컬럼을 인덱싱하세요.
df.iloc[:,[0]]
```

```python
# (5) 데이터프레임에서 iloc 속성을 사용하여 '037730' 종목의 '종목명', '현재가' 컬럼을 선택하세요.
df.iloc[[0],0:2]
```

```python
# (7) 데이터프레임에서 현재가가 1,500원 이상이면서 1,700원 미만인 값만 필터링 해보세요.
df[(df['현재가'] >= 1500) & (df['현재가'] < 1700)]
```

```python
# [문제2] 아래 코드를 활용해서 주어진 문제에 대한 코드를 작성하시오.
import pandas as pd
data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]

columns = ["테마", "종목명", "PER", "PBR"]
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
# (1) 데이터프레임에서 테마별 PER, PBR의 평균을 계산하세요.
df.groupby('테마').mean(numeric_only = True)
```

```python
# (2) 데이터프레임에서 테마별 종목의 개수를 계산하세요.
df.groupby('테마')[['종목명']].count()
df.groupby('테마').size()
```

```python
# (3) 데이터프레임에서 테마별 PER의 평균을 계산하세요.
df.groupby('테마')[['PER']].mean()
```

```python
# (4) 데이터프레임에 대해 테마로 그룹화한 후 PER, PBR 컬럼을 슬라이싱하세요. 그 다음 두 컬럼에 대한 평균을 계산하세요.
group = df.groupby('테마')
group[['PER','PBR']].mean()
```

```python
# (5) 데이터프레임에 대해서 '테마' 컬럼으로 그룹화한 후 PER 컬럼은 max, PBR 컬럼은 min 함수를 적용해보세요.
df.groupby('테마').agg({'PER':'max','PBR':'min'})
```

```python
# [문제3] 아래 코드를 활용해서 주어진 문제에 대한 코드를 작성하시오.
import pandas as pd

data1 = {
    'ID': [100, 101, 102, 103],
    'Name': ['Brayden', 'Clara', 'Ella', 'Oliver']
}

data2 = [
    [100, "삼성전자"],
    [100, "LG전자"],
    [101, "카카오"],
    [101, "NAVER"],
    [101, "현대차"]
]

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data=data2, columns=['ID', '종목명'])
```

```python
# (1) df1과 df2를 'ID'라는 컬럼명을 기준으로 'inner' 방식으로 병합하세요.
pd.merge(df1, df2, on = "ID")
```

```python
# (2) df1과 df2를 'ID’라는 컬럼명을 기준으로 ‘outer’ 방식으로 병합하세요.
pd.merge(df1, df2, on = "ID", how = "outer")
```

```python
# (3) 데이터프레임에서 df1의 'ID'와 df2의 '아이디' 컬럼을 기준으로 merge 하세요.(아래 코드 참조)
df2 = pd.DataFrame(data=data2, columns=['아이디', '종목명'])
pd.merge(df1, df2, left_on ='ID', right_on = '아이디')
```

```python
# [문제4] 다음 데이터프레임에서 df1, df2의 인덱스를 기준으로 merge 하는 코드를 작성하시오.
import pandas as pd

data1 = {
    'ID': [100, 101, 102, 103],
    'Name': ['Brayden', 'Clara', 'Ella', 'Oliver']
}
data2 = [
    [100, "삼성전자"],
    [100, "LG전자"],
    [101, "카카오"],
    [101, "NAVER"],
    [101, "현대차"]
]

df1 = pd.DataFrame(data1)
df1.set_index('ID', inplace=True)
df2 = pd.DataFrame(data=data2, columns=['ID', '종목명'])
df2.set_index('ID', inplace=True)
```

```python
pd.merge(df1, df2, left_index = True, right_index = True)
```
