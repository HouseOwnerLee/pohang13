```python
import pandas as pd
from pandas import Series
import numpy as np
```

```python
# 문제1
sample = Series([10,20,30,40,50],['가','나','다','라','마'])
print(sample)
```

```python
# 문제2
np.random.seed(20)
sample2 = pd.Series(np.random.randint(100,200,size =(15,)))
print(sample2[sample2 <= 160])
```

```python
# 문제3
s = Series(['apple',np.nan,'banana','kiwi','gubong'], ['가','나','다','라','마'])
print(s)
```

```python
# 문제4
print(s[s.notnull()])
```

```python
# 문제5
print(s[s.isna()].to_string())
```

```python
# 문제6
np.random.seed(0)
sample = pd.Series(np.random.randint(100, 200, size=(10,)))
print(sample[2:7])
```

```python
# 문제7
np.random.seed(0)
sample2 = pd.Series(np.random.randint(100, 200, size=(10,)), index=list('가나다라마바사아자차'))
print("(1)")
print(sample2['바':])
print()
print("(2))")
print(sample2[0:3])
print()
print("(3)")
print(sample2['나':'바'])
```
