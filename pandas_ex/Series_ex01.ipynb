```python
from pandas import Series

data = [1000, 2000, 3000]
index = ["메로나", "구구콘", "하겐다즈"]
s = Series(data,index)
print(s)
```

```python
print(s.iloc[0:2]) # iloc 번호로 접근
print(s.loc['메로나':'구구콘']) # loc 키값으로 접근
```

```python
indice = [0, 2]
print(s.iloc[indice])
print(s.iloc[[0,2]])
```

```python
indice = ["메로나", "하겐다즈"]
print(s.loc[indice])
print(s.loc[ ["메로나", "하겐다즈"]])
```

```python
s.loc['메로나'] = 500 # s.iloc[0] = 500. s['메로나'] = 500
print(s)
```

```python
s.loc['비비빅'] = 500 # 없는 키값인 경우 추가
print(s)
```

```python
s1 = s.drop('메로나')
print(s1)
```

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052] , ["서울", "부산", "인천", "대구"])
s2 = Series({"서울":9631482, "부산":3393191, "인천" : 2632035, "대전" : 1490158}, ["부산", "서울", "인천", "대전"])
ds = s - s2
print(ds)
```

```python
print(ds.notnull()) # 값이 null이면 False, 아니면 True
print(ds[ds.notnull()]) # null이 아닌 값만 출력함
```

```python
s3 = s / 1000000
print(s3)
```

```python
print(s.values - s2.values)
```

```python
rs = (s - s2) / s2 * 100
print(rs)
rs = rs[rs.notnull()]
print(rs)
```

```python
rs["부산"] = 1.63 # 부산 수정
rs["대구"] = 1.41 # 대구는 없으므로 추가됨
print(rs)
```

```python
del rs["서울"]
print(rs)
```

```python
from pandas import Series

date = ["6/1", "6/2", "6/3", "6/4", "6/5"]
high = Series([42800,42700,42050, 42950,43000], date)
low = Series([42150,42150,41300,42150,42350], date)
print(high)
print('---------------------')
print(low)
```

```python
diff = high - low
print(diff)
```

```python
print(diff.idxmax(), diff.max())
print(diff.idxmin(), diff.min())
```

```python
def remove_comma(x):
    return int(x.replace(",",""))
```

```python
s = Series(["1,234", "5,678", "9,876"])
result = list(s.map(remove_comma))
print(result)
```

```python
from pandas import Series

data = [42500, 42550, 41800, 42550, 42650]
index = ['2019-05-31', '2019-05-30', '2019-05-29', '2019-05-28', '2019-05-27']
s = Series(data,index)
print(s)
```

```python
from pandas import Series

data = [3.1, 2.0, 10.1, 5.1]
index = ['000010', '000020', '000030', '000040']
s = Series(data,index)
print(s)
```

```python
# s1 = s.sort_values() # 오름차순 정렬
s1 = s.sort_values(ascending = False) # 내림차순 정렬
print(s1)
```

```python
# s2 = s.rank() # 오름차순 기준으로 등수를 나눔
s2 = s.rank(ascending = False) # 내림차순으로 등수를 나눔
print(s2)
```
