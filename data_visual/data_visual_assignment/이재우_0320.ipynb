```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

(1) 팁 데이터에서 pivot_table()함수를 사용하여 index='day', columns='size', values='tip'으로하는 데이터를 만든 다음 해당 데이터로 히트맵을 작성하시오.

```python
sns.heatmap(tips.pivot_table(index='day', columns='size', values='tip').fillna(0),annot=True, cmap='inferno', fmt=".2f")
plt.title("Heatmap")
plt.show()
```

(2) 아이리스 데이터에서 각 데이터의 상관관계를 pairplot()으로 나타내시오.

```python
sns.pairplot(iris, diag_kind = "kde", hue="species", markers=["o", "s", "D"], height=3)
plt.suptitle("Iris Data의 Pair Plot", y=1.02)
plt.show()
```
