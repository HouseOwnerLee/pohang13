```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
tips.head()
```

```python
import matplotlib.pyplot as plt

sns.set_style("white") # darkgrid, whitegird, ticks, dark, white
sns.set_context("notebook") # paper, notebook, talk, poster
sns.set_palette('deep') # deep,, muted, pastel, dark, bright, colorblind
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h') # hue : time 속성으로 세분화한 그룹막대를 그림,orient : 수직, 수평 설정
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
x = iris.petal_length.values
plt.rcParams['font.family'] = 'malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

# 1차원 실수 분포 플롯

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
x = iris.petal_length.values
plt.rcParams['font.family'] = 'malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
sns.kdeplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
x = iris.petal_length.values
plt.rcParams['font.family'] = 'malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
sns.distplot(x, kde=True, rug=True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

# Countplot

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
sns.countplot(y="class", data=titanic, hue="who")
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.title("타이타닉호의 각 클래스별, 승객수")
plt.show()
```

# 다차원 데이터

## jointplot 
- 산점도 플롯과 함께 차트의 가장자리에 히스토그램을 그림

```python
import seaborn as sns

iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='kde', color='g') # kind : 그래프의 종류 (scatter, kde, hex)
plt.rcParams["font.family"] = "malgun Gothic"
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02) # y: 제목과의 간격
plt.show()
```

## pairplot
- 3차원 이상 다차원 실수형일때 사용
- 각 데이터 열의 조합에 대해 산점도 그림

```python
import seaborn as sns

iris = sns.load_dataset("iris")
sns.pairplot(iris, diag_kind = "kde", hue="species", markers=["o", "s", "D"], height=3)
plt.suptitle("Iris Data의 Pair Plot", y=1.02)
plt.show()
```

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head(2)
```

```python
titanic_size = titanic.pivot_table(index="class", columns="sex", aggfunc="size")
titanic_size
```

## heatmap
- 데이터가 2차원이고 모든 값이 카테고리 값일때

```python
sns.heatmap(titanic_size, cmap=sns.light_palette("gray", as_cmap=True), annot=True, fmt="d") # fmt: 나타낼 값의 형태
plt.title("Heatmap")
plt.show()
```

```python
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="YlGnBu",
           annot_kws={"size": 8, "weight" : "bold", "color" : "blue"}, fmt=".2f") # fmt : 소수점 3번째 자리에서 반올림
plt.show()
```

## barplot
- 카테고리 값에 따른 실수 값의 평균과 편차를 표시
- 두 가지 카테고리 값에 의한 실수값의 변화를 보기 위한 hue인수를 제공

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("요일 별, 전체 팁")
plt.show()
```

## boxplot


```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.boxplot(y="day", x="total_bill", data=tips, orient='h')
plt.title("요일 별 전체 팁의 Box Plot")
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Violin Plot")
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()
```

# 다차원 복합 데이터

```python
import seaborn as sns
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
titanic = sns.load_dataset("titanic")
data = titanic[titanic.survived.notnull()]
sns.catplot(x="age", y="sex", hue="survived", row="class", data=data,
           kind="violin", split=True, height=2,aspect=4)
plt.title("각 클래스의 성별 생존자 수의 Catplot")
plt.show()
```

```python
sns.catplot(x='who', y='survived',
            col='pclass',
            row='embarked',
            kind='bar',
            height=2.5,
            data=titanic)
plt.show()
```

```python
import seaborn as sns
import numpy as np

tips = sns.load_dataset("tips")
plt.title("Boxplot과 Strip Plot로 표현한, 요일 별 팁")
sns.boxplot(x="tip",y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip",y="day",data=tips,jitter=True, color="0.4")
plt.show()
```
