```python
f = open("data2.txt", "w")
f.close()

# 파일객체 = open(파일이름, 파일열기모드)

# r - 읽기모드: 파일을 읽기만 할때
# w - 쓰기모드: 파일의 내용을 쓸때
# a - 추가모드: 파일의 마지막에 내용을 추가할때
```

```python
f = open("./data/새파일.txt","w",encoding="utf-8")

for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)

f.close()
```

```python
f = open("data/새파일.txt", "r",encoding="UTF-8")
while True:
    line = f.readline()
    if not line: break
    print(line, end="")

f.close()
```

```python
# readlines() 함수 이용하기
f = open("data/새파일.txt","r",encoding="UTF-8")
lines = f.readlines()
print(lines)
f.close()
print(type(lines))
```

```python
# read 함수 사용하기
f = open("data/새파일.txt","r",encoding="utf8")
data = f.read()
print(data)
f.close()
```
