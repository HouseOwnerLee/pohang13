```python
# 문자열 인덱싱
a = "Life is too short"
print(a[8])
print(a[-1])
print(a[-2])
```

```python
# 문자열 슬라이싱
a = '''Life is too short, you need python'''

print(a[0:4])
print(a[12:17])
print(a[5:])
print(a[:17])
print(a[:])
```
