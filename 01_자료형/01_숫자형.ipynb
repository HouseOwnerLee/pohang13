```python
# 정수형

a = 123
b = -178
c = 0

print(a)
print(b)
print(c)
type(a)
```

```python
# 실수형

a = 1.2
b = -3.45

print(a)
print(b)
type(a)
```

```python
# 8진수, 16진수

a = 0o123
print(a)

b = 0x111
print(b)

0xF
```

```python
# 사칙연산

a = 3
b = 4

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
# x의 y제곱을 나타내는 **연산자

a = 3
b = 4

print(a**b)
```

```python
# 나눗셈 후 나머지를 구하는 % 연산자

print(7 % 3)
print(3 % 2)
```

```python
# 나눗셈 후 몫을 구하는 // 연산자

print(7 // 4)
```
